```python
#importing the libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #TF-Bow
from sklearn.feature_extraction.text import TfidfVectorizer #TFIDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    "My name is Chan Evan Wesley",
    "My first name is Evan",
    "My family name is Chan",
    "Wesley is my second first name",
    "I am pursuing my degree in ICT",
    "I love to eat fruits and ice cream",
    "I am a 21 years old young man"
]
```

```python
#Embeddings using TF-Bow approach
tf_vectorizer = CountVectorizer()
tf_matrix = tf_vectorizer.fit_transform(sentences)
```

```python
tf_matrix
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
tfidf_similarity = cosine_similarity(tfidf_matrix)
tfidf_similarity
```

```python
#finding out the document similarity
def doc_sim(similarity_matrix,sentences):
  for i in range(len(sentences)):
    sims=similarity_matrix[i].copy()
    sims[i] = -1
    top_index = np.argmax(sims)
    print(sentences[i], "-->",sentences[top_index])
```

```python
doc_sim(tfidf_similarity,sentences)
```

The similarity is based on lexical not in the meaning of the sentence
