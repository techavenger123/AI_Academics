```python
text2="""I am tony stark the real iron man who had killed thanos and his army, i had married pepper pots and invented time machine to meet my father in the past before i was born. . . . i am also the CEO and chairman of start industries and Head HR in Avenger to give salary. . . I am the king of New York. . ."""
```

**NER : Named Entity Recognition**

```python
for token in doc.ents:
    print(token.text,token.label_)
```

```python
for token in doc:
    print(token.text,"->",token.pos_)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
set(stopwords.words('english'))
```

```python
from nltk.stem.porter import *
```

```python
p_stem=PorterStemmer()
```

```python
t="""I studied the subject named ai and then meeting Mr. Rohan tomorrow in the meeting"""
```

```python
for word in t.split():
    print(word,"-->",p_stem.stem(word))
```

# **_Lemmatization_**

```python
for token in nlp(t):
    print(token.text,"-->",token.pos_,"->",token.lemma_)
```
